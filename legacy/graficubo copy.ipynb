```python
n = 60
d = 1/(n+1)

mid = int(1/(2*d))   # ≈ 128

i = j = k = mid

idx_i = lambda i: i + n*(j-1) + n**2*(k-1)
idx_j = lambda j: i + n*(j-1) + n**2*(k-1)
idx_k = lambda k: i + n*(j-1) + n**2*(k-1)

T_x = np.ones(n)
T_y = np.ones(n)
T_z = np.ones(n)

# (x, 1/2, 1/2)
for ii in range(n):
    T_x[ii] = T['T'].iloc[idx_i(ii+1) - 1]

# (1/2, y, 1/2)
for jj in range(n):
    T_y[jj] = T['T'].iloc[idx_j(jj+1) - 1]

# (1/2, 1/2, z)
for kk in range(n):
    T_z[kk] = T['T'].iloc[idx_k(kk+1) - 1]
```

```python
x = [i*d for i in range(1,n+1)]
```

```python
plt.plot(x,T_x,label= ' x')
plt.plot(x,T_y,label=  'y')
plt.plot(x,T_z,label= 'z')
plt.legend()
```
